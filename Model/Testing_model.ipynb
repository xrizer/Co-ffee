# Testing the model

```python
from tensorflow.keras.models import load_model
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
import numpy as np
```

Load the model

```python
model = load_model("Augmented_VGG16.h5")
```

Define the label

```python
label = {'Cerscospora': 0, 'Healthy': 1, 'Leaf rust': 2, 'Miner': 3, 'Phoma': 4}
lbl = list(label.keys())

def get_label(index):
   for class_string, class_index in label.items():
      if class_index == index:
         return class_string
```

Predicting images

```python
path="rps-results"

for img in os.listdir(path):
    img=image.load_img(path+"/"+img,target_size=(128,128))
    plt.imshow(img)
    plt.show()
    x=image.img_to_array(img)
    x=np.expand_dims(x,axis=0)
    images=np.vstack([x])
    classes = model.predict(images, batch_size=1)
    predicted_index = np.argmax(classes)
    pr=model.predict(images)[0]
    for(l,p) in zip(lbl,pr):
        print(f"{l} : {p*100:.2f}%")
    print(f"Hasil Prediksi : {get_label(predicted_index)}")
```
